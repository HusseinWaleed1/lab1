```python
# Import and get matplotlib ready
import matplotlib.pyplot as plt

# Prepare data (create two lists of 5 numbers, X & y)
X = [150043, 87899, 32549, 11179, 213095]   # Odometer
y = [4000, 5000, 7000, 22000, 3500]         # Price

# Setup figure and axes using plt.subplots()
fig, ax = plt.subplots()

# Add data (X, y) to axes
ax.plot(X, y, marker="o")

# Customize plot by adding a title, xlabel and ylabel
ax.set_title("Car Price vs Odometer")
ax.set_xlabel("Odometer (KM)")
ax.set_ylabel("Price (USD)")

# Save the plot to file using fig.savefig()
fig.savefig("car_price_vs_odometer.png")

plt.show()

```

```python
# Create a plot using plt.subplots() and plot X versus X^2 (X squared)
X = np.linspace(0, 100, 100)
y = X ** 2

fig, ax = plt.subplots()
ax.plot(X, y)

ax.set_title("X vs X^2")
ax.set_xlabel("X")
ax.set_ylabel("X squared")

plt.show()
```

We'll start with scatter plots.

```python
# Create a scatter plot of X versus the exponential of X (np.exp(X))
X = np.linspace(0, 5, 100)
y = np.exp(X)

fig, ax = plt.subplots()
ax.scatter(X, y)

ax.set_title("Scatter Plot of X vs exp(X)")
ax.set_xlabel("X")
ax.set_ylabel("exp(X)")

plt.show()
```

```python
# Create a scatter plot of X versus np.sin(X)
X = np.linspace(0, 100, 100)
y = np.sin(X)

fig, ax = plt.subplots()
ax.scatter(X, y)

ax.set_title("Scatter Plot of X vs sin(X)")
ax.set_xlabel("X")
ax.set_ylabel("sin(X)")

plt.show()
```

Now we've got a numeric column (`Total Sales`) and a dates column (`Sale Date`), let's visualize them.

```python
# Use the plot() function to plot the 'Sale Date' column versus the 'Total Sales' column
data = {
    "Sale Date": pd.date_range("2025-01-01", periods=10, freq="M"),
    "Total Sales": [5000, 7000, 6500, 8000, 9000, 7500, 8500, 9500, 10000, 11000]
}

car_sales = pd.DataFrame(data)

fig, ax = plt.subplots()
ax.plot(car_sales["Sale Date"], car_sales["Total Sales"], marker="o")

ax.set_title("Total Sales over Time")
ax.set_xlabel("Sale Date")
ax.set_ylabel("Total Sales")

plt.show()
```

```python
# Create a bar graph of the 'Make' and 'Odometer (KM)' columns in the car_sales DataFrame
data = {
    "Make": ["Toyota", "Honda", "Toyota", "BMW", "Nissan"],
    "Odometer (KM)": [150043, 87899, 32549, 11179, 213095],
    "Price": ["$4,000.00", "$5,000.00", "$7,000.00", "$22,000.00", "$3,500.00"]
}

car_sales = pd.DataFrame(data)

fig, ax = plt.subplots()
ax.bar(car_sales["Make"], car_sales["Odometer (KM)"])

ax.set_title("Car Make vs Odometer")
ax.set_xlabel("Make")
ax.set_ylabel("Odometer (KM)")

plt.show()
```
